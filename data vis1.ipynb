```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Replace null values of age by mean
# Replace numerical values by mean
age_mean=titanic['age'].mean()
```

```python
titanic['age'].fillna(value=age_mean,inplace=True)
```

```python
#Replace null values of deck by mode
# Deck means class->First class, second class
# Its non numeric
#hence replace by mode
deck_mode=titanic['deck'].mode()
```

```python
titanic['deck'].fillna(value=deck_mode,inplace=True)
```

```python
# Replace embarked town by mode
embarked_mode=titanic['embarked'].mode()
```

```python
titanic['embarked'].fillna(value=embarked_mode,inplace=True)
```

```python
#Replace embarkedtown with mode
embarked_mode=titanic['embark_town'].mode()
titanic['embark_town'].fillna(value=embarked_mode,inplace=True)
```

lmplot is a function in the Seaborn library that allows you to create a scatter plot with a linear regression line. The name "lmplot" stands for "linear model plot".

With lmplot, you can visualize the relationship between two variables and also explore how this relationship may vary across different subsets of your data.

```python
sns.lmplot(x='age', y='fare', data=titanic)
plt.xlabel('age')
plt.ylabel('fare')
plt.title('Scatter Plot with Linear Regression Line')

# Show the plot
plt.show()
```

sns.distplot() is a function in the seaborn library that can be used to create a histogram of a variable in a dataset. When you pass a variable to sns.distplot(), it creates a histogram of the distribution of that variable's values. The histogram shows the count or frequency of each value in the variable, grouped into bins of equal width.

In addition to the histogram, sns.distplot() can also plot a probability density function (PDF) curve, which is an estimate of the underlying probability distribution of the variable. By default, the PDF curve is plotted on top of the histogram, but you can set the kde parameter to False to only show the histogram and not the PDF.

```python
#Kde=true indicates-->Draw PDF curve
#KDE=false indicates ---> Dont draw PDF curve
```

The shape of the PDF curve can give you an idea of the  spread of the data. For example, a PDF that is symmetric and bell-shaped (like a normal distribution) indicates that the data is relatively evenly distributed around a central value, while a PDF that is skewed to the left or right indicates that the data is more concentrated in one direction than the other.

```python
sns.distplot(titanic['fare'], kde=True, bins=30)

# Add labels and a title
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Distribution of Ticket Prices')
```

Here's what each parameter in the sns.lmplot() function does:

x='age' specifies that the age variable should be plotted on the x-axis.
y='fare' specifies that the fare variable should be plotted on the y-axis.
data=titanic specifies the dataset that should be used for the plot.
fit_reg=False specifies that a regression line should not be fit to the data points.
hue='sex' specifies that the plot should be segmented by the sex variable in the dataset.
markers=['o', 'x'] specifies that the markers used for male and female passengers should be circles and crosses, respectively.
height=5 specifies the height of the plot in inches.

```python
# Create a new column 'age_group' to categorize passengers by age
titanic['age_group'] = pd.cut(titanic['age'], bins=[0, 18, 30, 50, 100], labels=['Child', 'Young Adult', 'Middle Aged', 'Elderly'])

# Calculate the survival rate for each age group
#-We have created groups and we have to find survival rate of each group
#-so select a group
#-find mean of survived column in that group
#Then save that data into a new dataframe
'''

Why to use reset_index()?

see when we have divided data into saperate groups the original index given to
data will be scattered.
Eg : Before grouping index of "a" is 10. Lets assume a is the youngest person
After grouping, a will be placed in first row,in younger group (As a is most youngest person present)
But the original index 10 will remain same
hence reset index to 0


'''

age_group_survival = titanic.groupby('age_group')['survived'].mean().reset_index()

# Create a bar plot of the survival rate by age group
sns.barplot(x='age_group', y='survived', data=age_group_survival)

# Set the title of the bar plot
plt.title('Survival Rate by Age Group')

# Set the x-axis label
plt.xlabel('Age Group')

# Set the y-axis label
plt.ylabel('Survival Rate')

# Show the plot
plt.show()
```

```python
'''
Cross-tabulation, also known as contingency table or crosstab, is a statistical method 
to analyze the relationship between two categorical variables. 
It helps to understand how often different categories appear together or independently from one another.

For example, let's say we have data on the gender and employment status of a group of people. 
We can create a cross-tabulation table to see how many people are employed and unemployed in each gender
category. This will show us the distribution of employment status within each gender group.


Gender	Employed	Unemployed
Male	       100	50
Female	      70	30

'''
```

```python
# Create a cross tabulation table of age group and survival
survival_table = pd.crosstab(titanic['age_group'], titanic['survived'], values=titanic['survived'], aggfunc='count')

# Print the cross tabulation table
print(survival_table)
```

```python
'''
In this case, the 'values' parameter is set to 'titanic['survived']' 
which means that the count of passengers who survived or did not survive
 will be displayed in the table. The 'aggfunc' parameter is set to 'count'
  which means that the counts of the values will be shown in the table.

'''
```

```python
# Rename the column labels
survival_table.columns = ['Did not survive', 'Survived']
# Print the cross tabulation table
print(survival_table)
```

```python


# Create a bar plot of the survival count by age group
sns.barplot(x=survival_table.index, y='Survived', data=survival_table)

# Set the title of the bar plot
plt.title('Survival Count by Age Group')

# Set the x-axis label
plt.xlabel('Age Group')

# Set the y-axis label
plt.ylabel('Survival Count')

# Show the plot
plt.show()
```

In this code, we first create a bar plot of the survival count by age group using the 'sns.barplot' function. We set the x-axis to be the index of the 'survival_table' dataframe which contains the age group as the index. We set the y-axis to be the 'Survived' column of the 'survival_table' dataframe which contains the count of passengers who survived in each age group. We set the color to be green and label it as 'Survived'.

Next, we create a bar plot of the death count by age group using the same 'sns.barplot' function. We set the x-axis to be the index of the 'survival_table' dataframe and the y-axis to be the 'Did not survive' column of the 'survival_table' dataframe which contains the count of passengers who did not survive in each age group. We set the color to be red and label it as 'Did not survive'.

We set the bottom parameter of the second 'sns.barplot' function to be the 'Survived' column of the 'survival_table' dataframe, which stacks the death count bars on top of the survival count bars.

Finally, we set the title of the plot as 'Survival and Death Count by Age Group', the x-axis label as 'Age Group', and the y-axis label as 'Count'. We add a legend to differentiate between the survival count and death count bars. We then show the plot using 'plt.show()'.

```python
# Create a bar plot of the survival count by age group
# x axis=index of survival table
# y axis="survived column" of survival table
sns.barplot(x=survival_table.index, y='Survived', data=survival_table, color='green', label='Survived')

# Create a bar plot of the death count by age group
# x axis=index of survival table
# y axis="did not servived" column of survival table
#Bottom=>This bar plot should have bottom as survived
'''
In the sns.barplot() function, the bottom parameter is used to specify the y-value(s)
at which the bar(s) should start. In this case, bottom=survival_table['Survived'] 
means that the "Did not survive" bars will be stacked on top of the "Survived" bars, 
starting at the height of the "Survived" bars.

'''
sns.barplot(x=survival_table.index, y='Did not survive', data=survival_table, color='red', label='Did not survive', bottom=survival_table['Survived'])

# Set the title of the bar plot
plt.title('Survival and Death Count by Age Group')

# Set the x-axis label
plt.xlabel('Age Group')

# Set the y-axis label
plt.ylabel('Count')

# Show the legend
plt.legend()

# Show the plot
plt.show()    
```
