Perform Min Max Normalization:
Min-max normalization is a data scaling technique that rescales numeric features to a range between 0 and 1

x_norm = (x - min(x)) / (max(x) - min(x))


```python
# normalize column 'ANTI_A'
normalized_A = (group1['ANTI_A'] - group1['ANTI_A'].min()) / (group1['ANTI_A'].max() - group1['ANTI_A'].min())

# replace original column with normalized column
group1['ANTI_A'] = normalized_A

# normalize column 'ANTI_B'
normalized_B = (group1['ANTI_B'] - group1['ANTI_B'].min()) / (group1['ANTI_B'].max() - group1['ANTI_B'].min())

# replace original column with normalized column
group1['ANTI_B'] = normalized_B

print(group1)
```

Theory
A scatter plot is a type of chart that shows the relationship between two variables by plotting them as points on a two-dimensional graph. It's called a scatter plot because the points scatter around the graph.

For example, let's say we want to compare the weight and height of a group of people. We can create a scatter plot with height on the y-axis and weight on the x-axis. Each person's height and weight will be plotted as a point on the graph. The position of the point will depend on the height and weight of the person.

By looking at the scatter plot, we can see if there's a relationship between height and weight. If the points on the plot are spread out and don't form a pattern, then there's probably no relationship. But if the points form a straight line or some other recognizable pattern, then there might be a relationship between the two variables.

Scatter plots are useful for identifying trends and patterns in data, especially when working with large datasets. They can also be used to identify outliers or unusual observations in the data.

In summary, a scatter plot is a visual representation of the relationship between two variables, with each point on the plot representing an observation in the dataset. It's a useful tool for identifying patterns and trends in data.

```python
import matplotlib.pyplot as plt


plt.scatter(group1['ANTI_A'], group1['ANTI_B'])

# add labels and title
plt.xlabel('ANTI_A')
plt.ylabel('ANTI_B')
plt.title('Scatter Plot GROUP 1')

# show plot
plt.show()
```

Pearson correlation:
The Pearson correlation coefficient is a statistical measure that shows the degree of linear relationship between two variables. It is widely used in statistical analysis to quantify the strength and direction of the relationship between two variables.

The Pearson correlation coefficient can be used to:

Determine whether two variables are positively or negatively correlated: A positive correlation means that as one variable increases, the other variable also tends to increase. A negative correlation means that as one variable increases, the other variable tends to decrease.

```python
# calculate pearson correlation coefficient
corr_matrix = group1.corr(method='pearson')
pearson_coef = corr_matrix.loc['ANTI_A', 'ANTI_B']

print("Pearson correlation coefficient:", pearson_coef)
```
